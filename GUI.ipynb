```python
from PIL import ImageTk, Image, ImageDraw, ImageOps
import PIL
import tkinter as tk
from tkinter import *
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

#defining canvas details
width = 500
height = 500
white = (255, 255, 255)





#########################Next bit transfers over some stuff from the neural_net_training.py##################################

#each label is encoded, so we also read in the file emnist-balanced-mapping.txt for decoding purposes
mapping = pd.read_table('emnist-balanced-mapping.txt', names = ['label', 'unicode'], header = None, delim_whitespace  = True)

#column 0 represents the labels in the data, and column 1 is the corresponding unicode integer. 
#we can see what character each label represents using the chr() function
char_vec = []
for i in range(len(mapping)):
    a = chr(mapping.loc[i][1])
    char_vec.append(a)
mapping['character'] = char_vec

#loading pre-trained hyperparameters
W1 = np.load( "W1_pretrained.dat" , allow_pickle=True)
W2 = np.load( "W2_pretrained.dat" , allow_pickle=True)
B1 = np.load( "B1_pretrained.dat" , allow_pickle=True)
B2 = np.load( "B2_pretrained.dat" , allow_pickle=True)

#redefining foward net
#ReLU activation and its derivative
def ReLu(x):
    return x*(x>0)
#softmax activation function
def softmax(x):
    z = x - np.amax(x)
    return np.exp(z)/np.sum(np.exp(z))
#define forward propagation
def forward(W_in, b_in, W_hid, b_hid, x):
    
    #layer 1
    Z1 = W_in.dot(x) + b_in #weights and biases
    A1 = ReLu(Z1) #activation
    
    #layer 2
    Z2 = W_hid.dot(A1) + b_hid #weights and biases
    A2 = softmax(Z2) #activation

    return Z1, A1, Z2, A2

############################################################################################################################







#function that draws using cursor
def paint(event):
    x1, y1 = (event.x - 10), (event.y - 10)
    x2, y2 = (event.x + 10), (event.y + 10)
    cv.create_line(x1, y1, x2, y2, fill="black",width=40)
    draw.line([x1, y1, x2, y2],fill="black",width=40)
    
#function that saves mouse drawn image and passes it through neural net and  prediction
def model():
    
    filename = "image.png"
    #resizes the image to 28 by 28 pixel format as this is the format the neural network trained with
    new_image = image1.resize((28, 28))
    #converts image to grayscale and inverts black and white. Again because of the format the neural network trained with
    im_invert = ImageOps.invert(new_image)
    im_invert = im_invert. convert('L')
    #saves the drawn image
    im_invert.save(filename)
    #convert image to numpy 
    vec  = np.asarray(im_invert)
    #transpose and reshape to needed format. Lastly rescale by factor of 255, again due to training choices in neuralnet_training.py
    vec = np.transpose(vec)
    vec = vec.reshape((784,1))
    vec = vec/255
    
    #we reshape the vector into a 28 by 28 pixel format for plotting purposes
    image_Example = vec.reshape((28,28))
    #we transpose this new matrix to ensure the image is the right way up
    image_Example = np.transpose(image_Example)
    #we plot the image
    plt.gray()
    plt.imshow(image_Example)
    plt.show()
    
    
    
    #forward passing image through neural network
    Z1, A1, Z2, A2 = forward(W1, B1, W2, B2, vec)
    #prediction
    prediction = np.argmax(A2)
    
    #returning prediction and confidence 
    label1.config(text = "Prediction: " + chr(mapping.loc[prediction][1]))
    label2.config(text = "Prediction confidence: " + str(100*A2[prediction,0] ) + "%")
    print("Prediction: " + chr(mapping.loc[prediction][1]))
    print("Prediction confidence: " + str(100*A2[prediction,0] ) )

#function to clear canvas and prediction    
def clear():
    cv.delete('all')
    draw.rectangle((0, 0, 500, 500), fill=(255, 255, 255, 0))
    #clears label
    label1.config(text = "Prediction: ")
    label2.config(text = "Prediction confidence: ")

#creating window and canvas    
root = Tk()
root.resizable(0,0)
cv = Canvas(root, width=width, height=height, bg='white')
cv.pack()

# PIL create an empty image and draw object to draw on
# memory only, not visible
image1 = PIL.Image.new("RGB", (width, height), white)
draw = ImageDraw.Draw(image1)

#pack canvas
cv.pack(expand=YES, fill=BOTH)
#binds mouse to drawing fucntion
cv.bind("<B1-Motion>", paint)

#empty labels which is filled with the prediction when predict button is pressed
label1 = Label(root, text= "Prediction: " )
label2 = Label(root, text= "Prediction confidence: "  )
#this adds the new labels to the GUI
label1.pack()
label2.pack()

#Predict and clear buttons
btnModel=Button(text="Predict",command=model)
btnClear=Button(text="clear",command=clear)
btnModel.pack()
btnClear.pack()

#title
root.title('Slowly draw a letter/number and click on predict')
root.mainloop()
```
